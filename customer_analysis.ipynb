```python

df1 = pd.read_csv('/content/ecommerce_customer_data_custom_ratios.csv')
df2 = pd.read_csv('/content/ecommerce_customer_data_large.csv')
```

```python
# First few rows
print(df1.head())
print(df2.head())

# Info
df1.info()
df2.info()

# Describe stats
print(df1.describe())
print(df2.describe())

# Missing values
print(df1.isnull().sum())
print(df2.isnull().sum())
```

```python
# Remove duplicates
df1.drop_duplicates(inplace=True)
df2.drop_duplicates(inplace=True)

# Fill missing values with 0
df1.fillna(0, inplace=True)
df2.fillna(0, inplace=True)
```

```python
plt.figure(figsize=(8,5))
plt.hist(df1['Customer Age'], alpha=0.5, label='Custom Ratios')
plt.hist(df2['Customer Age'], alpha=0.5, label='Large Dataset')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Customer Age Comparison')
plt.legend()
plt.show()
```
